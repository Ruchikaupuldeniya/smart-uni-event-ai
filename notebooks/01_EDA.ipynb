මේ Notebook එකෙන් අපි බලන්නේ අපේ University Event Recommender එකට හදාගත්ත Data වල තියෙන Patterns මොනවද කියලා.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set aesthetic parameters for seaborn
sns.set_theme(style="whitegrid")

# Load Data
DATA_DIR = '../data'
students_df = pd.read_csv(os.path.join(DATA_DIR, 'students.csv'))
events_df = pd.read_csv(os.path.join(DATA_DIR, 'events.csv'))
attendance_df = pd.read_csv(os.path.join(DATA_DIR, 'attendance.csv'))

print(f"Students: {len(students_df)}")
print(f"Events: {len(events_df)}")
print(f"Attendance Records: {len(attendance_df)}")
```

### 1. ළමයි වැඩිපුරම ඉන්නේ මොන University වලද?

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=students_df, y='university', order=students_df['university'].value_counts().index, palette='viridis')
plt.title('Distribution of Students Across Universities')
plt.xlabel('Number of Students')
plt.ylabel('University')
plt.show()
```

### 2. ගොඩක්ම ජනප්‍රිය Event Categories මොනවද?

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=events_df, x='category', order=events_df['category'].value_counts().index, palette='magma')
plt.title('Most Popular Event Categories')
plt.xlabel('Category')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()
```

### 3. Attendance Rate එක (RSVP කරපු අයගෙන් ඇත්තටම ආපු ගාණ)

```python
plt.figure(figsize=(6, 6))
attendance_counts = attendance_df['attended'].value_counts()
plt.pie(attendance_counts, labels=['Attended (1)', 'Did Not Attend (0)'], autopct='%1.1f%%', colors=['#4CAF50', '#F44336'], startangle=90)
plt.title('Event Attendance Rate')
plt.show()
```

### 4. ළමයි ගොඩක්ම කැමති Interests (Tags) මොනවද?

```python
# Split the comma-separated interests and count them
all_interests = students_df['interests'].str.split(', ').explode()
interest_counts = all_interests.value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=interest_counts.values, y=interest_counts.index, palette='cubehelix')
plt.title('Top 10 Most Popular Student Interests')
plt.xlabel('Number of Students')
plt.ylabel('Interest Area')
plt.show()
```
